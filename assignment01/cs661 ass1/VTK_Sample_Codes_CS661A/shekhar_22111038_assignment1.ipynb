### 1.1 Data Query/Processing Task:

```python

## Import VTK
from vtk import *

## Load data
#######################################
reader = vtkXMLImageDataReader()
reader.SetFileName('Data/Isabel_2D.vti')
reader.Update()
data = reader.GetOutput()

## Query how many cells the dataset has
#######################################
numCells = data.GetNumberOfCells()
print("Number of cells in the dataset = ",numCells)
d = data.GetDimensions()
print("Dimensions of the dataset = ",d)
x,y,z = data.GetDimensions()
noOfPoints = x*y*z
print("Number of points = ", noOfPoints)
rng = data.GetScalarRange()
print("Range of data = ", rng)

dataArr = data.GetPointData().GetArray('Pressure')
print("average pressure of entire dataset:")
sum=0
for i in range(noOfPoints):
    sum += dataArr.GetTuple1(i)
print(sum/noOfPoints)

## Get a single cell from the list of cells
###########################################
cellId = 0;
cell = data.GetCell(cellId) ## cell index = 0

## Query the 4 corner points of the cell
#########################################
pid1 = cell.GetPointId(0)
pid2 = cell.GetPointId(1)
pid3 = cell.GetPointId(3)
pid4 = cell.GetPointId(2)

## Print the 1D indices of the corner points
############################################
print('1D indices of the cell corner points:')
print(pid1,pid2,pid3,pid4) ## in counter-clockwise order



## Print the locations (3D coordinates) of the points
#######################################################
print('Point locations of cell corners in counter clockwise order:')
cor1 = data.GetPoint(pid1)
cor2 = data.GetPoint(pid2)
cor3 = data.GetPoint(pid3)
cor4 = data.GetPoint(pid4)
print(cor1)
print(cor2)
print(cor3)
print(cor4)

print("3D coordinates of the cell center:")
xcent = (cor1[0]+cor2[0]+cor3[0]+cor4[0])/4
ycent = (cor1[1]+cor2[1]+cor3[1]+cor4[1])/4
zcent = (cor1[2]+cor2[2]+cor3[2]+cor4[2])/4
print(xcent,ycent,zcent)

## Get values at each vertex
## First Get the array
print("Pressure values of the corner points of the cell:")

val1 = dataArr.GetTuple1(pid1)
val2 = dataArr.GetTuple1(pid2)
val3 = dataArr.GetTuple1(pid3)
val4 = dataArr.GetTuple1(pid4)
print(val1,val2,val3,val4)



print("Average pressure value at the center of the cell:")
print((val1+val2+val3+val4)/4)
```

### 1.2 Visualization Task:

```python
## Extract and store one cell
#############################
points = vtkPoints()
points.InsertNextPoint(data.GetPoint(pid1))
points.InsertNextPoint(data.GetPoint(pid2))
points.InsertNextPoint(data.GetPoint(pid3))
points.InsertNextPoint(data.GetPoint(pid4))

### Create a polydata
######################
pdata = vtkPolyData()

### Add points to polydata
####################################
pdata.SetPoints(points)

# setup colors
# colors = vtkNamedColors()
colors = vtkUnsignedCharArray()
colors.SetNumberOfComponents(3)
colors.SetName('Colors')
colors.InsertNextTuple3(255,0,0)
colors.InsertNextTuple3(0,255,0)
colors.InsertNextTuple3(0,0,255)
colors.InsertNextTuple3(100,100,100)

pdata.GetPointData().SetScalars(colors)
pdata.Modified()

# Create a vtkVertexGlyphFilter object
glyphFilter = vtkVertexGlyphFilter()
glyphFilter.SetInputData(pdata)
glyphFilter.Update()

# Get the output of the filter
glyphData = glyphFilter.GetOutput()

# Create a vtkPolyDataMapper to map the output to the graphics library
mapper = vtkPolyDataMapper()
mapper.SetInputData(glyphData)

# Create a vtkActor to represent the output in a render window
actor = vtkActor()
actor.SetMapper(mapper)
actor.GetProperty().SetPointSize(30)

### Setup render window, renderer, and interactor
##################################################
renderer = vtkRenderer()
renderer.SetBackground(1,1,1)
renderWindow = vtkRenderWindow()
renderWindow.SetSize(800,800)
renderWindow.AddRenderer(renderer)
renderWindowInteractor = vtkRenderWindowInteractor()
renderWindowInteractor.SetRenderWindow(renderWindow)
renderer.AddActor(actor)


### Finally render the object
#############################
renderWindow.Render()
renderWindowInteractor.Start()
```
